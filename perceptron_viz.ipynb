# Perceptron Algorithm

In this Jupyter Notebook, our goal is to add a real-time animation of the perceptron separating our input data

As You may notice, this implementation isn't quite useful in a real aplication (as a library, anyway). However, You'll be able to visualize the perceptron converging with ease.

```python
# First, importing essential libraries that we will use to manipulate vectors and create graphs
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.animation as animation

```

```python

def init_weights(lower_boun,upper_bound):
    w0 = random.randint(lower_boun,upper_bound) #Initializing weights at random
    w1 = random.randint(lower_boun,upper_bound)
    w2 = random.randint(lower_boun,upper_bound)

    weights = np.array([w0,w1,w2]).astype('float')
    return weights


def signal(x):
    if x>=0:
        return 1
    else:
        return -1
```

```python
learning_rate = 1 # determines how big a step we take when adjusting weights
weights = init_weights(-9,9) #initalizing weights at random

# below, we manually add the equivalent of an 'AND' gate
# this will make more sense as we get further into the notebook 
data_points = np.array([[1,0,1], [1,1,0], [1,1,1], [1,0,0]]).astype('float')
classes = [-1,-1,1,-1]
```

```python
%matplotlib notebook

fig = plt.figure()
p1 = fig.add_subplot(111)
i=0
def plot_line(k):#calculates ax+b function]
    # for i in range(len(data_points)):
    global i, weights, data_points
    signal_output = signal(np.dot(data_points[i],weights))
    if signal_output==classes[i]:
        pass
        # counter = counter+1
    else:
        if signal_output==-1:
            weights = weights+(data_points[i])*learning_rate # multiply by learning rate
        else:
            weights = weights-(data_points[i])*learning_rate # multiply by learning rate

    print(weights)

    i = (i+1)%4

    p1.clear() # pra limpar a linha anterior
    p1.scatter([1], [1], color='red')
    p1.scatter([1,0,0], [0,1,0], color='green')
    x = np.linspace(-5,5,100)
    y = (-weights[0]-(weights[1]*x))/weights[2]
    plt.title('Perceptron')
    plt.xlabel('X1', color='#1C2833')
    plt.ylabel('X2', color='#1C2833')
    plt.xlim(-0.5, 2)
    plt.ylim(-0.5, 2)
    plt.grid()
    p1.plot(x,y)

    
graph = animation.FuncAnimation(fig, plot_line, interval=100)
plt.show()
```
